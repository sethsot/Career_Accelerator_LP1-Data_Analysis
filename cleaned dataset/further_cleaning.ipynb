```python
#Top 10 preferred industries
Top_10_industries=df.groupby('Sector')['Amount($)'].sum().reset_index()
Top_10_industries=Top_10_industries.sort_values(by='Amount($)', ascending=False)
Top_10_industries.head(10)
```

```python

# Format 'Amount($)' column to two decimal points
Top_10_industries['Amount($)'] = Top_10_industries['Amount($)'].apply(lambda x: '{:.2f}'.format(x))

# Create a style dictionary to bold the column names
style_dict = {
    'Amount($)': 'font-weight: bold',
    'Sector': 'font-weight: bold'
}

# Apply the style to the DataFrame
styled_table = Top_10_industries.style.set_table_styles([style_dict])

# Display the styled table
display(styled_table)
```

```python
# Top 10 preferred industries
Top_10_industries = df.groupby('Sector')['Amount($)'].sum().reset_index()
Top_10_industries = Top_10_industries.sort_values(by='Amount($)', ascending=False).head(10)

# Drop the index
Top_10_industries.reset_index(drop=True, inplace=True)

# Reduce zeros in the 'Amount($)' column
Top_10_industries['Amount($)'] = Top_10_industries['Amount($)'] / 1000000  # Divide by 1,000,000 to reduce zeros

# Create a style dictionary to bold the column names
style_dict = {
    'Amount($)': 'font-weight: bold',
    'Sector': 'font-weight: bold'
}

# Print the top 10 industries
print(Top_10_industries)
```

```python
# replace similar categories in stage column
df['Stage']=df['Stage'].replace('Unknown Stage','Unknown')

# Filter the data to exclude rows with 'unknown' stage
filtered_df = df[df['Stage'] != 'Unknown']

# Compute the count of each stage
stage_count = filtered_df['Stage'].value_counts().head(10)

# Sort the stage_count in descending order
stage_count = stage_count.sort_values(ascending=True)

# Create the plot
plt.figure(figsize=(10, 5))
plt.yticks(rotation=0)
plt.barh(stage_count.index, stage_count.values, align='center')
plt.title('Stage Distribution')
plt.show()
```

```python
# Filter the data to exclude rows with 'unknown' stage
filtered_df = df[df['Stage'] != 'Unknown']

# Compute the count of each stage
stage_count = filtered_df['Stage'].value_counts().head()

# Sort the stage_count in descending order
stage_count = stage_count.sort_values(ascending=True)

# Create the horizontal bar plot
plt.figure(figsize=(10, 5))
plt.yticks(rotation=0)
bars = plt.barh(stage_count.index, stage_count.values, align='center')

# Add frequencies on each bar
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, bar.get_width(), ha='left', va='center')

# Remove x-axis tick labels
plt.xticks([])

# Remove x-label and y-label
plt.xlabel('')
plt.ylabel('')

# Set the title
plt.title('Top 5 Stages of Funding')

# Show the plot
plt.show()
```

```python

# Get the top locations by count
top_locations = df['HeadQuarter'].value_counts().head().index

# Filter the data to include only the top locations
filtered_data = df[df['HeadQuarter'].isin(top_locations)]

# Group the filtered data by 'HeadQuarter' and 'Sector' and calculate the count
sector_location_count = filtered_data.groupby(['HeadQuarter', 'Sector']).size().unstack()

# Create the stacked column chart
sector_location_count.plot(kind='bar', stacked=True, figsize=(10, 5))

# Set the title and labels
plt.title('Sectors in Top 5 Headquarters')
plt.xlabel('HeadQuarter')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python


# Calculate the count of each 'HeadQuarter'
headquarter_count = df['HeadQuarter'].value_counts().head(4)

# Get the corresponding 'Amount($)' for each 'HeadQuarter'
headquarter_amount = df.groupby('HeadQuarter')['Amount($)'].sum()

# Create a DataFrame with the top 4 'HeadQuarter' and their corresponding count and amount
bubble_data = pd.DataFrame({'HeadQuarter': headquarter_count.index,
                            'Count': headquarter_count.values,
                            'Amount($)': headquarter_amount[headquarter_count.index].values})

# Create the bubble plot
plt.figure(figsize=(35, 25))
plt.scatter(bubble_data['HeadQuarter'], bubble_data['Count'], s=bubble_data['Amount($)'] / 1000000, alpha=0.7)

# Set labels and title
plt.xlabel('HeadQuarter')
plt.ylabel('Count')
plt.title('Bubble Plot of Top 4 HeadQuarters')

# Show the plot
plt.show()
```
